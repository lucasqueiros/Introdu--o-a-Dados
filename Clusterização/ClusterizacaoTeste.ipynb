###Criando colunas dummy
Agora com os dados tratados e numericos

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

criando numero de clusters

```python
n_clusters = 4
```

```python
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(df_optativas)
```

```python
labels = kmeans.labels_
```

```python
df_optativas['cluster'] = labels
```

```python
cluster_stats = df_optativas.groupby('cluster').mean()
cluster_stats
```

```python
scaler = StandardScaler()
df_standardized = scaler.fit_transform(df_optativas)
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_standardized)
df_pca = pd.DataFrame(data=pca_result, columns=['Componente1', 'Componente2'])
```

```python
plt.scatter(df_pca['Componente1'], df_pca['Componente2'], c=labels, cmap='viridis')
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.title('Resultado do PCA')
plt.show()
```
