```python
import pandas as pd

urlDados = 'https://raw.githubusercontent.com/elasComputacao/raio-x-dados/main/data/dados-processados/idade.csv'
dados = pd.read_csv(urlDados)
```

```python
dados.head() #exibe as 5 primeiras linhas da sua tabela
dados.tail() #exibe as 5 ultimas linhas da sua tabela
display(dados) #exibe uma visao geral com as primeiras e ultimas linhas da tabela
```

```python
dados['idade_ingresso'].value_counts().plot.bar()
#primeiro passo - 'idade_ingresso' define qual coluna de dados vamos plotar
#segundo passo - .valuecounts() calcula quantas vezes cada dado se repete
#terceiro passo - .plot.bar() plota um grafico de barras, esse metodo nao precisa de importação de biblioteca
```
