```python
#cria um dicionario
dados = {'Estado' : ['Santa Caratina', 'Rio de Janeiro', 'Paraiba', 'Pernambuco', 'Bahia', 'Minas Gerais'],
				'Ano' : [2004, 1997, 1808, 2023, 2006, 2019], 'Taxa Desemprego' : [1.5, 6.2, 0.1, 2.5, 1.6, 0.9]}
```

```python
#imprimir o dicionario
print(dados)
```

```python
#importa a função DataFrame do Pandas
from pandas import DataFrame
```

```python
#converte o dicionario em um dataframe do pandas
df = DataFrame(dados)
```

```python
#visualiza as 5 primeiras linhas
df.head()
```

```python
#Reorganizando as colunas
DataFrame(dados, columns = ['Estado', 'Taxa Desemprego', 'Ano'])
```

```python
#criando outro dataFrame com os mesmos dados anteriores mas adicionando uma coluna
df2 = DataFrame(dados,
								columns = ['Estado', 'Taxa Desemprego', 'Taxa Crescimento', 'Ano'],
								index = ['estado1', 'estado2', 'estado3', 'estado4', 'estado5', 'estado6'])
```
